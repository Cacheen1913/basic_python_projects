```python
# UNIT CONVERTER

print("Welcome User!")

while True:
    user = int(input("Here are the list of conversion which you can perform\n"
                "1. Kilometer to Meter\n"
                "2. Dollar to INR\n"
                "3. Hours to Seconds\n"))
    print("Press 0 for exit")

    if user == 1:
        #kilometer to meter
        k =float(input("ENTER THE DISTANCE IN KILOMETRES : "))
        m = (k * 1000)
        print("Your distance in meter: ", m, "m")

    elif user == 2:
        #dollar to inr
        d = float(input("ENTER THE AMOUNT IN DOLLARS : "))
        INR = (88.5*d)
        print("Your AMOUNT in INR : Rs", INR)

    elif user == 3:
        #hours to seconds
        h = int(input("ENTER THE TIME IN HOUR : "))
        SECONDS = (h*3600)
        print("Your TIME in seconds : ", SECONDS, "s")

    elif user == 0:
        print("Bye Bye!")
        break


```
