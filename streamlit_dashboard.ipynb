```python
#DASHBOARD

import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

st.title("Streamlit app of vgu")
st.text("welcome to my dashboard")
st.header("I am header")

name = st.text_input("enter your name:\n")
age = st.number_input("enter your age:")

st.write("your name is:", name , "your age is:", age)
st.selectbox("Enter Your Choice:",["BTECH","MTECH","BCA","BSC"])

if st.button("Submit"):
    st.success("Registered Successfully..")
file = st.file_uploader("Upload your file")

if file:
    content = file.read()
    st.write("File Uploaded Successful...")

data = {"Name": ["Debojit","Sachin","Ayush"], "Marks": [90,87,70] }
df = pd.DataFrame(data)

st.dataframe(df)

data = pd.DataFrame({
  "Marks":  [90,87,70]
})

st.line_chart(data)
st.bar_chart(data)
st.area_chart(data)

subject = ["python","c++","java"]
marks = [75,50,90]

fig , ax  = plt.subplots()
ax.pie(marks,labels=subject,autopct='%1.1f%%')

st.pyplot(fig)
```
