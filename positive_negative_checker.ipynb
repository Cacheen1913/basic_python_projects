```python
#POSITIVE AND NEGATIVE NUMBER CHECKER

while True:
    print("0 FOR EXIT")
    num = int(input("Enter a integer to check whether it is positive or negative: "))
    if num == 0:
        print("Bye Bye")
        break
    elif num < 0:
        print("The number is negative.")
    elif num > 0:
        print("The number is positive.")
        
```
